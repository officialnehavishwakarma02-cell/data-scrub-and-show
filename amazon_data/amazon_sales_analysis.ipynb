#Amazon Sales Analysis:Data Loading and Initial Inspection
In this section,I am setting up the environment and importing the raw dataset.
### Libraries
**Pandas**:Used for data manipulation and analysis
Dataset Source 
data is sourced from an excel file containing raw amazon sales transaction,including Order IDs,Category information,and customer details.

```python
import pandas as pd
df=pd.read_excel("C:\\Users\\conta\\Downloads\\archive\\Amazon 2_Raw.xlsx")
print(df)
```

### Data Inspection
Now that the data is loaded into a pandas Dataframe,i am checking the first few rows to understand the column 
structure any missing values or privacy concerns like(emailid column that needs to be addresses before analysis)

```python
df=pd.DataFrame(df)
df
```

I am renaming the columns to **snake_case**(lowercase with underscores)to improve the developmet workflow
**Reasons for this step:**
* **programming Ease:**Accessing 'df.order_id' is faster and less error-prone than d'f[Order ID']'.
* **Readability:** it creates a clean,professional look for the final dataset.

```python
df=df.rename(columns={'Order ID':'order_id',
                      'Order Date':'order_date',
                      'Ship Date':'ship_date',
                      'EmailID':'email_id',
                      'Geography':'geography',
                      'Category':'category',
                      'Product Name':'product_name',
                      'Sales':'sales',
                      'Quantity':'quantity',
                      'Profit':'profit'})
df
```

```python
df.head()
```

# Final Step:Remove Personally Identifiable Information(PII)

```python
df=df.drop_duplicates(['email_id','order_id'])
df
```

# The professional way to split columns and avoid warnings

# Final Step:Remove Personally Identifiable Information

```python
df=df.drop('geography',axis=1)
df
```

#Display the final,clean,anonymized table  

```python
df.head()
```

### Financial Data Polishing:Rounding Sales and Profit
To Ensure Data is ready for business reporting, i am rounding the numerical values to two decimal places

```python
df[['sales','profit']]=df[['sales','profit']].round(2)
df
```

### Data Normalization:Remove Leading and Trailing spaces
I am using the '.str.strip()' method on the 'category' column to ensure there are no accidental spaces or end of the text strings

```python
df['category']=df['category'].str.strip()
df
```

### Finalizing the Dataset:Index Reset
After Cleaning the data and removing duplicate records,the row indices were no longer sequential.I am resetting the index to provide a clean,continuous 
numerical order

```python
df=df.reset_index(drop=True)
df
```

###Final Data Anonymization
To Finalize the dataset for public analysis, I am removing all personally identifiable information

```python
if 'email_id' in df.columns:
    df=df.drop(columns=['email_id'])
    df.head()
```

```python
df.to_csv('data/cleaned_amazon_sales.csv',index=False)
```

### Sales Analsis by Category
I am visualizing the total sales for each product category to identify the top-performing segmetns.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
if not os.path.exists('neha'):
    os.makedirs('neha')
plt.figure(figsize=(12,8))
#grouping data
df_sorted=df.groupby('category')['sales'].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='sales',y='category',data=df_sorted,palette='viridis',hue='category',legend=False)
plt.title("total sales by category")
plt.xlabel('total sales',fontsize=12)
plt.ylabel('category',fontsize=12)
plt.savefig('neha/total_sales_category.png',dpi=300,bbox_inches='tight')
plt.show()
```

###Monthly Revenue Performance (Month-End)
This line chart visualizes the total sales revenue aggregated at the end  of each month
**key obersavtion:**
* **Trend Analysis:** By grouping data by month-end we can see the periodic fluctuations in Amazon's sales performance
* **Data Smoothing:** Resampling by month-end helps remove the 'noise' of daily fluctuations to show a clearer business trajectory

```python
sales_overtime=df.resample('ME',on='order_date')['sales'].sum()
sales_overtime
```

```python
import matplotlib.pyplot as plt
monthly_trend=df.resample('ME',on='order_date')['sales'].sum()
monthly_trend.plot(figsize=(12,5),color='RED',marker='o')
plt.title('Monthly Sales Trend(2011-2014)')
plt.ylabel('Total Revenue')
plt.grid(True,alpha=0.3)
plt.show()
```

### Relationship Analysis:Sales vs Quantity
This scatter plot examines the correlation between the number of items ordered and the total sale value for each trasanction,categorized by product segment
**What this graph reveals:**
* **Volume vs Value:**We can identify whether high sales are driven by high-priced individual items or large quantities of cheaper products
* **Category Distribution:** by usinf 'hue='category',we can see which product types tend to have larger bulk orders versus one-off high-value purchases

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='quantity',y='sales',hue='category',alpha=0.3)
plt.title('individual order:sales vs quantity')
plt.savefig('visualizations/individual_orders.png',dpi=300,bbox_inches='tight')
plt.show()
```

#Inventory Volume:Total Quantity Sold per category
In this section,I a analyzing which product categories have the highest turnover by calculatingthe total units sold(quantity) for each
* **why this matters for business:**
* **DeMAND Forecasting:** knowing that 'bonders' and 'paper' sell in massive quantity helps with warehouse space management

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,5))
sorted_data=df.groupby('category')['quantity'].sum().reset_index()
sorted_data=sorted_data.sort_values('quantity',ascending=False)
sns.barplot(x='quantity',y='category',data=sorted_data)
plt.title("quantity per category")
plt.xlabel('quantity')
plt.ylabel('category')
plt.savefig('visualizations/quantity_per_category.png',dpi=300,bbox_inches='tight')
plt.show()
```

### Top 10 High-Revenue Products
This visulization identifies the top 10  products that generated the highest total sales revenue
**Strategic Business Insights:**
**Revenue Drivers:** We can pinpoint exactly which specific items(like the Global Troy Executive Leather Chair or the hewlett Packered Laserjet)are the 
biggest contributor to the bottom line
**Marketing Focus:** These products represents the most successful items in the catalog,making them primary candidates for featured promotions or inventory prioritizaion

```python
sk=df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
sk
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
sns.barplot(x='sales',y='product_name',data=sk.reset_index(),hue='product_name',palette='viridis')
plt.title("top 10 selling product",fontsize=20)
plt.xlabel("Sales",fontsize=18)
plt.ylabel('product_name',fontsize=18)
plt.savefig('visualizations/top_10_selling_product.png',dpi=300,bbox_inches='tight')
plt.show()
```

###Sales Analysis by state
We are grouping the data by state to visualize total sales performance across different regions:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sales_by_country=df.groupby('state')['sales'].sum().reset_index()
plt.figure(figsize=(12,5))
sns.barplot(x='sales',y='state',data=sales_by_country,hue='state')
plt.title('sales by state')
plt.xlabel("sales")
plt.ylabel("state")
plt.savefig('visualizations/sales_by_state.png',dpi=300,bbox_inches='tight')
plt.show()
```

### Sales Contribution Analysis
The following visualization displays the percentage of sales contributed by each category,sorted from highest to lowest:

```python
category_sales=df.groupby('category')['sales'].sum()
sales_percentage=(category_sales/category_sales.sum()*100)
sales_percentage
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sales_percentage.sort_values(ascending=False).plot(kind='bar')
plt.title("Sales Contribution Percentage by Category")
plt.xlabel('Sales Contributuon(%)')
plt.ylabel("Category")
plt.savefig('visualizations/sales_contribution_percentage_category.png',dpi=300,bbox_inches='tight')
plt.show()
```
